```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
data=np.array([
    [1,150000,160000,170000,180000,190000,200000],
    [2,155000,165000,175000,185000,195000,205000],
    [3,160000,170000,180000,190000,200000,210000],
    [4,165000,175000,185000,195000,205000,215000],
    [5,170000,180000,190000,200000,210000,220000]])

print("=== Zomato data analysis ===")
print("\n sales data shape:", data.shape)
#print("\n sales data:\n", data)
print("First 3 restaturant sales data:\n", data[0:3])
```

```python
# total per year
print(np.sum(data, axis=0))
print("Total sales per year:",np.sum(data[:,1:], axis=0))
# minimum per restaurant
print("Minimum sales per restaurant:",np.min(data[:,1:],axis=1))
# maxminium per year
print("Maximun sales per year:",np.amax(data[:,1:],axis=0))
# average per restaurant
print("Average sales per restaurant:",np.mean(data[:,1:], axis=1))
```

```python
# cumulative sales per restaurant
cumsum=np.cumsum(data[:,1:], axis=1)
print("Cumulative sales per restaurant:\n", cumsum)

plt.figure(figsize=(10,6))
plt.plot(np.mean(cumsum, axis=0))
plt.title("Average Cumulative Sales Over Years")
plt.xlabel("Year")
plt.ylabel("Sales")
plt.grid(True)
plt.show()
```
