## Student Data

```python
import numpy as np
```

### Genrating student data

```python
arr=np.random.randint(0, 101, (500, 5))
print(arr)
```

### Calculating the total marks of each student

```python
total_marks = np.sum(arr, axis=1)
print("Total Marks:", total_marks)
```

```python
#Student percentage calculation
percentage=(total_marks/500)*100
for i, p in enumerate(percentage[:500], start=1):
    print(f"Student {i}: {p:.2f}%")
```

```python
#Student Grading System
conditions = [
    percentage >= 90,
    percentage >= 85,
    percentage >= 80,
    percentage >= 75,
    percentage >= 70,
    percentage >= 65,
    percentage >= 60,
    percentage >= 50
]

grades = ['A+','A','B+','B','C+','C','D+','D']

Grade = np.select(conditions, grades, default='F')

print(Grade)
```

```python
#Falied students
failed_students = np.where(Grade == 'F')[0] + 1
print("Failed Student Numbers:", failed_students)
print("Total Failed Students:", len(failed_students))
```

```python
#Topper student
topper_student= np.argmax(total_marks)
print(f"Topper Student Number: {topper_student + 1}")
print("Topper student marks:", total_marks[topper_student])
print("Topper student percentage:",percentage[topper_student])
```

```python
#Subject-wise toppers
subject_toppers = np.argmax(arr, axis=0)

print("Subject Toppers:")
for i in range(5):
    print(f"Subject {i+1}: Student {subject_toppers[i]+1}")
```

```python
#class statics
print("Highest Marks:", np.max(total_marks))
print("Lowest Marks:", np.min(total_marks))
print("Average Marks:", np.mean(total_marks))
print("Median Marks:", np.median(total_marks))
print("Standard Deviation:", np.std(total_marks))
```

```python
#Number of students in each grade
unique, counts = np.unique(Grade, return_counts=True)

for grade, count in zip(unique, counts):
    print(f"{grade}: {count}")
```

```python
#top 10 students
ranks = np.argsort(total_marks)[::-1]

print("Top 10 Students:")
for i in range(10):
    student = ranks[i] + 1
    marks = total_marks[ranks[i]]

    print(f"Rank {i+1}: Student {student} - {marks} marks")
```
